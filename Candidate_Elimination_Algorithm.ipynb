```python
import pandas as pd
import numpy as np

# -------------------------------
# Create the training dataset
# -------------------------------
data = {
    'Sky': ['Sunny', 'Sunny', 'Rainy', 'Sunny'],
    'AirTemp': ['Warm', 'Warm', 'Cold', 'Warm'],
    'Humidity': ['Normal', 'High', 'High', 'High'],
    'Wind': ['Strong', 'Strong', 'Strong', 'Strong'],
    'Water': ['Warm', 'Warm', 'Warm', 'Cool'],
    'Forecast': ['Same', 'Same', 'Change', 'Change'],
    'EnjoySport': ['Yes', 'Yes', 'No', 'Yes']
}

df = pd.DataFrame(data)

concepts = np.array(df.iloc[:, :-1])
target = np.array(df.iloc[:, -1])

# -------------------------------
# Candidate Elimination Algorithm
# -------------------------------
S = concepts[0].copy()
G = [['?' for _ in range(len(S))] for _ in range(len(S))]

for i, instance in enumerate(concepts):

    if target[i] == 'Yes':
        for j in range(len(S)):
            if instance[j] != S[j]:
                S[j] = '?'
                G[j][j] = '?'

    elif target[i] == 'No':
        for j in range(len(S)):
            if instance[j] != S[j]:
                G[j][j] = S[j]
            else:
                G[j][j] = '?'

# Remove all rows containing only '?'
G = [g for g in G if g != ['?'] * len(S)]

# -------------------------------
# Display Final Result
# -------------------------------
print("\nResult\n")
print("The Candidate-Elimination algorithm gives:\n")

print("Specific Boundary S:")
print(list(S))

print("\nGeneral Boundary G:")
print(G)
```
